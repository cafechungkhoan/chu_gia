```python
import pandas as pd
df = pd.read_excel('Book1.xlsx')
```

```python
import pandas as pd
import plotly.graph_objects as go

# Count the transitions between 'Qualified Lead (Y/ N)' and 'Call Result'
qualified_result_counts = df.groupby(['Qualified Lead (Y/ N)', 'Call Result']).size().reset_index(name='count')

# Create mappings for 'Qualified Lead' categories and 'Call Result' categories to numeric indices
qualified_mapping = {'Y': 0, 'N': 1, 'U': 2}
call_result_mapping = {result: idx + 3 for idx, result in enumerate(df['Call Result'].unique())}

# Build the Sankey data
sources = []
targets = []
values = []
colors = []  # Custom color for each link

for _, row in qualified_result_counts.iterrows():
    sources.append(qualified_mapping[row['Qualified Lead (Y/ N)']])
    targets.append(call_result_mapping[row['Call Result']])
    values.append(row['count'])

    # Assign green to 'Successful Policy', others to a default color
    if row['Call Result'] == 'Successful Policy':
        colors.append('rgba(0, 255, 0, 0.8)')  # Green with some transparency
    else:
        colors.append('rgba(100, 100, 100, 0.5)')  # Gray with some transparency

# Labels for each part of the Sankey diagram
labels = ['Qualified: Yes', 'Qualified: No', 'Qualified: Unknown'] + list(call_result_mapping.keys())

# Create the Sankey diagram using Plotly
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=colors  # Apply custom colors to links
    )))

# Update layout and render the Sankey chart
fig.update_layout(title_text="Sankey Diagram: Qualified Leads and Call Results (Highlighting Successful Policy)", font_size=10)
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Count occurrences of 'Qualified Lead' and 'Location' pairs
qualified_location_counts = df.groupby(['Qualified Lead (Y/ N)', 'Location']).size().reset_index(name='count')

# Create mappings for 'Qualified Lead' categories and 'Location' categories to numeric indices
qualified_mapping = {'Y': 0, 'N': 1, 'U': 2}
location_mapping = {location: idx + 3 for idx, location in enumerate(df['Location'].unique())}

# Build the Sankey data
sources = []
targets = []
values = []
colors = []  # To store colors for each flow

for _, row in qualified_location_counts.iterrows():
    sources.append(qualified_mapping[row['Qualified Lead (Y/ N)']])
    targets.append(location_mapping[row['Location']])
    values.append(row['count'])

    # Apply green color for "Qualified: Yes", gray for others
    if row['Qualified Lead (Y/ N)'] == 'Y':
        colors.append('rgba(0, 255, 0, 0.6)')  # Green with some transparency
    elif row['Qualified Lead (Y/ N)'] == 'N':
        colors.append('rgba(255, 0, 0, 0.6)')  # Red for "No"
    else:
        colors.append('rgba(100, 100, 100, 0.5)')  # Gray for "Unknown"

# Labels for each part of the Sankey diagram
labels = ['Qualified: Yes', 'Qualified: No', 'Qualified: Unknown'] + list(location_mapping.keys())

# Create the Sankey diagram using Plotly
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=colors  # Apply custom colors to links
    )))

# Update layout and render the Sankey chart
fig.update_layout(title_text="Sankey Diagram: Qualified Leads and Locations (Green for Yes)", font_size=10)
fig.show()
```
